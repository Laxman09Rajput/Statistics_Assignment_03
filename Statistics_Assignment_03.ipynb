# Statistics

1. What is hypothesis testing in statistics ?

 - A statistical technique known as hypothesis testing is used to determine whether a claim about a population is likely to be true based on sample data. It starts with two hypotheses: the null hypothesis (H₀), which assumes that there is no effect or difference, and the alternative hypothesis (H₁), which suggests that there is an effect or difference. Data is collected and analyzed using tests such as t-tests, chi-square tests, or ANOVA. The p-value helps determine whether to reject H₀; if the p-value is below a chosen significance level (e.g., 0.05), we reject H₀; if not, we fail to reject H₀. Hypothesis testing offers an organized method of making decisions using data, reducing uncertainty and aiding in fields such as science, business, and medicine.



 2. What is the null hypothesis, and how does it differ from the alternative hypothesis

 - The null hypothesis in statistics is the notion that there is no significant difference or that nothing has changed in the scenario under study.  It's our starting position; it's like claiming a new medication doesn't function any better than the old one.  On the other hand, the alternative hypothesis asserts that there is a difference or effect.  By examining data, researchers test the null hypothesis to see whether there is sufficient evidence to refute it.  They embrace the alternative and reject the null if the evidence clearly indicates otherwise.  To put it simply, the alternative indicates that "something is happening," whereas the null indicates that "nothing is happening."  In their research, scientists and researchers are better able to make well-informed conclusions thanks to this method of thinking.

3. What is the significance level in hypothesis testing, and why is it important  

  - The significance level in hypothesis testing, usually represented by alpha (α), is the probability of rejecting a true null hypothesis. In simpler terms, it’s the risk you’re willing to take of making a wrong decision when the null hypothesis is actually correct. Common significance levels are 0.05 (5%) or 0.01 (1%). This value sets the threshold for determining whether the result of a test is statistically significant. If the p-value is less than α, we reject the null hypothesis, suggesting that the observed result is unlikely to have occurred by chance. The significance level is important because it helps control the chance of making false conclusions, especially in scientific research where accurate decisions matter. Choosing the right α ensures balanced judgment between being too cautious and too risky


4.  What does a P-value represent in hypothesis testing ?

   - In hypothesis testing, the p-value represents the probability of getting results as extreme as the ones observed, assuming that the null hypothesis is true. It helps us measure the strength of the evidence against the null hypothesis. A smaller p-value means stronger evidence to reject the null hypothesis, while a larger p-value suggests that the observed result could have happened by random chance. For example, if the p-value is 0.03, there is a 3% chance the results are due to random variation, assuming the null hypothesis is correct. The p-value is compared to the significance level (like 0.05) to make decisions. If the p-value is less than the significance level, we reject the null hypothesis. This makes the p-value a crucial tool for drawing conclusions from data.

   5. How do you interpret the P-value in hypothesis testing

    - Interpreting the p-value in hypothesis testing is about understanding how likely your results are if the null hypothesis is actually true. A low p-value (typically less than 0.05) means that the observed results are unlikely to have occurred by chance, so we usually reject the null hypothesis. This suggests that there may be a real effect or difference. On the other hand, a high p-value (greater than 0.05) means that the results could easily occur by random chance, so we fail to reject the null hypothesis and assume there’s no strong evidence of a real effect. It's important to remember that the p-value doesn’t prove a hypothesis; it just shows how consistent the data is with the assumption that the null hypothesis is true


   6. What are Type 1 and Type 2 errors in hypothesis testing

      - In hypothesis testing, Type 1 and Type 2 errors are mistakes we might make when making decisions based on sample data. A Type 1 error happens when we reject a true null hypothesis, meaning we say there is an effect or difference when there actually isn’t. It’s like a false alarm—believing something has changed when it hasn’t. The probability of making a Type 1 error is called the significance level (alpha). A Type 2 error, on the other hand, happens when we fail to reject a false null hypothesis, meaning we miss a real effect or difference. It’s like ignoring a fire alarm when there really is a fire. The probability of making a Type 2 error is called beta. Balancing both errors is key to making reliable decisions.

   7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing

   - In hypothesis testing, the main difference between a one-tailed and a two-tailed test lies in the direction of the hypothesis. A one-tailed test is used when we are only interested in checking if the effect or difference is in one specific direction—either greater than or less than a certain value. For example, if we want to test whether a new medicine works better than the current one, we use a one-tailed test. In contrast, a two-tailed test is used when we are interested in finding any difference, regardless of direction—whether it's higher or lower. This is common when we just want to know if there’s a change at all. Two-tailed tests are more conservative because they check for significance on both ends of the distribution, while one-tailed tests focus on only one side.

   8. What is the Z-test, and when is it used in hypothesis testing

    - The Z-test is a statistical method used in hypothesis testing to determine whether there is a significant difference between sample and population means, or between two sample means, when the population standard deviation is known. It is most useful when working with large sample sizes (usually n > 30) and when the data follows a normal distribution. The Z-test helps us understand if the difference we observe is due to chance or if it's statistically meaningful. For example, if a company claims the average height of its employees is 170 cm, we can use a Z-test to check if a sample of employees supports this claim. It’s called a "Z-test" because the results are compared to the standard normal distribution (Z-distribution). It’s a common test in fields like business, healthcare, and social science.

   9. How do you calculate the Z-score, and what does it represent in hypothesis testing

   - The Z-score is a value that tells us how many standard deviations a data point is from the mean. In hypothesis testing, it helps us measure how far our sample result is from the expected value under the null hypothesis. To calculate the Z-score, we use the formula:
   
    Z = (X̄ - μ) / (σ / √n)
  
   where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size. The Z-score allows us to compare our result to the standard normal distribution and find the probability (p-value) of getting that result by chance. A high or low Z-score means the sample result is far from the expected mean, which might lead us to reject the null hypothesis if the p-value is small enough.


   10. What is the T-distribution, and when should it be used instead of the normal distribution
   
   - The T-distribution is a type of probability distribution used in statistics when dealing with small sample sizes (usually less than 30) and when the population standard deviation is unknown. It looks similar to the normal distribution but has thicker tails, which means it gives more room for variation. This helps to provide more accurate results when sample data is limited or less precise. The T-distribution becomes more like the normal distribution as the sample size increases. You should use the T-distribution instead of the normal (Z) distribution when you're working with a small sample and only have the sample standard deviation, not the population's. It’s commonly used in t-tests to check for differences between means when full population data isn’t available, making it very important in real-world research situations.

   11. What is the difference between a Z-test and a T-test ?

    - The main difference between a Z-test and a T-test lies in the sample size and whether the population standard deviation is known. A Z-test is used when the sample size is large (typically over 30) and the population standard deviation is known. It relies on the normal distribution. On the other hand, a T-test is used when the sample size is small (less than 30) and the population standard deviation is unknown. It uses the T-distribution, which is more spread out to account for extra uncertainty in small samples. Both tests help compare means and check if results are statistically significant, but the T-test is more flexible for real-world situations where complete population data is usually not available. As sample size increases, both tests give similar results.

   12. What is the T-test, and how is it used in hypothesis testing ?

    - The T-test is a statistical tool used in hypothesis testing to determine if there is a significant difference between the means of two groups or between a sample mean and a known value. It’s especially helpful when working with small samples and when the population standard deviation is unknown. The T-test compares the difference between group averages relative to the variation in the data, helping us understand if the difference is likely due to chance or if it’s meaningful. For example, a researcher might use a T-test to check if a new teaching method improves student scores compared to the old one. By calculating the T-value and comparing it to a critical value from the T-distribution, we decide whether to reject or accept the null hypothesis, guiding conclusions in experiments and studies.

   13.  What is the relationship between Z-test and T-test in hypothesis testing ?

    - The Z-test and T-test are both used in hypothesis testing to compare means and check if differences are statistically significant, but they are connected by when and how they’re applied. The Z-test is used when the sample size is large and the population standard deviation is known, relying on the normal distribution. The T-test is used when the sample size is small and the population standard deviation is unknown, using the T-distribution, which accounts for extra uncertainty. As the sample size grows larger, the T-distribution starts to look like the normal distribution, so the T-test results become similar to the Z-test. In this way, the T-test can be seen as a more flexible version of the Z-test, suitable for real-world situations where population information is limited or sample sizes are small.

    14. What is a confidence interval, and how is it used to interpret statistical results ?

    - A confidence interval is a range of values that gives us an estimated range where the true population parameter, like a mean or proportion, is likely to fall. Instead of giving a single number, it provides a range that reflects the uncertainty in our estimate. For example, a 95% confidence interval means we are 95% confident that the true value lies within that range. It helps us understand how precise our estimate is—the narrower the interval, the more precise the estimate. Confidence intervals are important because they give more information than just a yes-or-no answer. They show the possible variation in results and help us interpret how reliable our findings are in hypothesis testing and research, guiding better decision-making.

    15. What is the margin of error, and how does it affect the confidence interval ?

     - The margin of error is the amount of uncertainty or possible error in a statistical estimate, like when you measure an average from a sample instead of the whole population. It tells us how much the estimate might differ from the true value. The margin of error directly affects the confidence interval by determining its width—the bigger the margin of error, the wider the confidence interval will be. For example, if the margin of error is small, we have a narrower range and more precise estimate. If it’s large, the confidence interval is wider, showing less certainty about the true value. Factors like sample size and variability influence the margin of error. Understanding it helps us know how confident we can be in our results and how reliable our conclusions are.

     16.  How is Bayes' Theorem used in statistics, and what is its significance ?

     - Bayes' Theorem is a powerful tool in statistics that helps us update our beliefs or predictions based on new evidence. It combines prior knowledge (what we already know) with new data to give us a better, updated probability of something happening. For example, doctors use it to update the chances of a disease being present after seeing a test result. It’s written as

     P(A|B) = [P(B|A) × P(A)] / P(B)

  This means the probability of event A given B has happened is based on how likely B is if A is true, times how likely A was before, divided by how likely B is overall. Bayes' Theorem is significant because it supports decision-making under uncertainty and is widely used in fields like medicine, machine learning, finance, and data science.

  17. What is the Chi-square distribution, and when is it used ?

    - The Chi-square distribution is a special type of probability distribution used mainly for testing relationships between variables and checking how well data fits a particular model. It’s commonly used in Chi-square tests, such as the goodness-of-fit test and the test for independence in contingency tables. For example, if a researcher wants to know whether there is a relationship between gender and product preference, they can use the Chi-square test. The Chi-square distribution is skewed to the right and depends on the degrees of freedom, which are based on the number of categories or variables involved. It is only used with categorical (non-numeric) data. The larger the Chi-square value, the more likely it is that the observed results are significantly different from what we would expect by chance, helping us make informed decisions.

  18. What is the Chi-square goodness of fit test, and how is it applied ?

  - The Chi-square goodness of fit test is a statistical method used to see if the observed data matches what we would expect based on a specific hypothesis. It’s helpful when we want to check whether a set of observed frequencies (counts) fits a particular pattern or distribution. For example, if we roll a die 60 times and expect each number (1 to 6) to appear 10 times, we can use this test to check if the actual results are close enough to this expectation. We calculate the Chi-square statistic by comparing the observed and expected values, then check if the difference is big enough to be considered significant. If it is, we reject the null hypothesis, meaning the data doesn’t follow the expected pattern. This test is commonly used in quality control, surveys, and research studies.

  19. What is the F-distribution, and when is it used in hypothesis testing ?

    - The F-distribution is a type of probability distribution used mainly in hypothesis testing when comparing two variances or in ANOVA (Analysis of Variance) tests. It helps us find out whether the differences between group means are statistically significant. The F-distribution is skewed to the right and is defined by two different degrees of freedom—one for the numerator and one for the denominator. It is used when we want to test if multiple groups have the same variance or if at least one group is different. For example, in ANOVA, it helps determine if students from different classes perform differently in exams. If the calculated F-value is larger than the critical value from the F-distribution table, we reject the null hypothesis, suggesting that at least one group differs significantly from the others.

 20. What is an ANOVA test, and what are its assumptions ?

    - The ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to see if there’s a significant difference between them. Instead of doing multiple t-tests, which increases error risk, ANOVA looks at all groups together using a single test. For example, if a teacher wants to compare test scores from students in three different classes, ANOVA can tell if the performance differences are meaningful or just due to chance. However, ANOVA relies on a few important assumptions: the data should be normally distributed, the variances of all groups should be equal (homogeneity of variance), and the samples should be independent of each other. If these assumptions are met, ANOVA gives reliable results and helps in making better, evidence-based decisions in research.
     

  21. What are the different types of ANOVA tests ?
  
   - There are several types of ANOVA tests designed to address different research scenarios. The one-way ANOVA is used when comparing the means of three or more independent groups based on a single factor. In contrast, the two-way ANOVA assesses the effect of two independent factors simultaneously, and it also explores any interaction between these factors. When the same subjects are measured under different conditions or at different times, a repeated measures ANOVA is appropriate, as it accounts for the correlation between measurements from the same individuals. Additionally, a multivariate ANOVA (MANOVA) extends the analysis to include multiple dependent variables, offering a broader view of the differences among groups. Each type helps tailor the analysis to the specific design and data structure of a study.

22. What is the F-test, and how does it relate to hypothesis testing ?

   - The F-test is a statistical method used in hypothesis testing to compare variances between two or more groups. It helps us determine if the differences in data spread (variability) are significant or just due to chance. The F-test is most commonly used in ANOVA (Analysis of Variance) to see if the means of multiple groups are significantly different. It does this by comparing the variance between groups to the variance within groups. If the F-value is high, it suggests that at least one group mean is different from the others. This test is important in situations like comparing student performance across different teaching methods or treatments. It’s called the F-test because it follows the F-distribution, and it plays a key role in identifying real effects in experiments or studies.

   


```python
##Practical

import numpy as np
from scipy.stats import norm

population_mean = 100
population_std = 15

np.random.seed(0)
sample = np.random.normal(loc=105, scale=15, size=50)


sample_mean = np.mean(sample)


z = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))

p = 2 * (1 - norm.cdf(abs(z)))


print("Sample Mean:", round(sample_mean, 2))
print("Z-score:", round(z, 2))
print("P-value:", round(p, 4))

if p < 0.05:
    print("✅ The difference is statistically significant.")
else:
    print("❌ The difference is NOT statistically significant.")
```

```python
import numpy as np
from scipy.stats import ttest_1samp


np.random.seed(0)
sample = np.random.normal(100, 10, 30)


t_value, p_value = ttest_1samp(sample, 105)


print("Sample average:", round(np.mean(sample), 2))
print("T-value:", round(t_value, 2))
print("P-value:", round(p_value, 4))


if p_value < 0.05:
    print("Result: Sample is significantly different from 105.")
else:
    print("Result: No significant difference from 105.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0=50, mu1=53, sigma=5, n=30, alpha=0.05):
    """
    mu0: population mean under H0
    mu1: population mean under H1
    sigma: standard deviation
    n: sample size
    alpha: significance level
    """


    standard_error = sigma / np.sqrt(n)
    z_critical = norm.ppf(1 - alpha)
    critical_value = mu0 + z_critical * standard_error

    x = np.linspace(40, 60, 1000)
    y0 = norm.pdf(x, mu0, standard_error)
    y1 = norm.pdf(x, mu1, standard_error)

    plt.figure(figsize=(10, 6))


    plt.plot(x, y0, label="Null Hypothesis (H0)", color='blue')
    plt.plot(x, y1, label="Alternative Hypothesis (H1)", color='green')


    plt.fill_between(x, 0, y0, where=(x > critical_value), color='red', alpha=0.4, label="Type I Error (α)")

    plt.fill_between(x, 0, y1, where=(x <= critical_value), color='orange', alpha=0.4, label="Type II Error (β)")

    plt.axvline(critical_value, color='black', linestyle='--', label=f"Critical Value = {round(critical_value, 2)}")

    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid(True)
    plt.show()


visualize_type1_type2()
```

```python
import numpy as np
from scipy.stats import t


def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_dev = np.std(sample, ddof=1)
    standard_error = std_dev / np.sqrt(n)


    t_critical = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * standard_error


    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return round(lower, 2), round(upper, 2), round(mean, 2)

sample_data = [78, 82, 85, 88, 79, 81, 83, 86, 84, 80]

low, high, avg = confidence_interval(sample_data)

print(f"Sample Mean: {avg}")
print(f"95% Confidence Interval: ({low}, {high})")
```

```python
import math
from scipy.stats import norm

def margin_of_error(data, confidence_level):

    n = len(data)


    mean = sum(data) / n

    variance = sum((x - mean) ** 2 for x in data) / (n - 1)
    std_dev = math.sqrt(variance)


    z = norm.ppf(1 - (1 - confidence_level) / 2)

    margin_error = z * (std_dev / math.sqrt(n))

    return margin_error


sample_data = [20, 22, 19, 24, 20, 21, 23, 22, 20, 21]
confidence = 0.95

error = margin_of_error(sample_data, confidence)
print(f"Margin of Error at {confidence*100}% confidence: {error:.2f}")
```

```python
def bayes_theorem(prior, likelihood, false_positive_rate):


    p_positive = likelihood * prior + false_positive_rate * (1 - prior)

    posterior = (likelihood * prior) / p_positive
    return posterior

prior_disease = 0.01
likelihood_positive_given_disease = 0.99
false_positive_rate = 0.05

prob_disease_given_positive = bayes_theorem(prior_disease, likelihood_positive_given_disease, false_positive_rate)
print(f"Probability of having disease given a positive test: {prob_disease_given_positive:.4f}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = {'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male'],
        'Preference': ['Tea', 'Coffee', 'Tea', 'Coffee', 'Tea', 'Tea', 'Coffee', 'Coffee']}

df = pd.DataFrame(data)

contingency_table = pd.crosstab(df['Gender'], df['Preference'])

chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)

if p_value < 0.05:
    print("There is a significant relationship between Gender and Preference.")
else:
    print("There is no significant relationship between Gender and Preference.")
```

```python
import numpy as np
from scipy.stats import f

sample1 = [25, 30, 28, 35, 32, 31, 29]
sample2 = [22, 27, 26, 30, 24, 23, 25]

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

if var1 > var2:
    f_stat = var1 / var2
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

p_value = 1 - f.cdf(f_stat, dfn, dfd)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Variances are significantly different.")
else:
    print("No significant difference in variances.")
```

```python
from scipy.stats import f_oneway

group1 = [23, 20, 25, 30, 22]
group2 = [35, 40, 38, 32, 37]
group3 = [45, 47, 50, 46, 48]

f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference between the group means.")
else:
    print("There is no significant difference between the group means.")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

group1 = [23, 20, 25, 30, 22]
group2 = [35, 40, 38, 32, 37]
group3 = [45, 47, 50, 46, 48]

f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference between the group means.")
else:
    print("There is no significant difference between the group means.")

data = [group1, group2, group3]

plt.boxplot(data, labels=["Group 1", "Group 2", "Group 3"])
plt.title("Group Comparison using ANOVA")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Score': [90, 85, 88, 75, 78, 82, 93, 95, 97, 65, 60, 58],
    'TeachingMethod': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
    'StudentType': ['Beginner', 'Advanced', 'Beginner', 'Advanced',
                    'Beginner', 'Advanced', 'Beginner', 'Advanced',
                    'Beginner', 'Advanced', 'Beginner', 'Advanced']
})

model = ols('Score ~ C(TeachingMethod) + C(StudentType) + C(TeachingMethod):C(StudentType)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

sns.pointplot(data=data, x='TeachingMethod', y='Score', hue='StudentType', capsize=.1)
plt.title("Interaction between Teaching Method and Student Type")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

x = np.linspace(0.01, 5, 500)

dfn = 5  # degrees of freedom numerator
dfd = 10  # degrees of freedom denominator

y = f.pdf(x, dfn, dfd)

plt.plot(x, y, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='blue')
plt.fill_between(x, y, alpha=0.2, color='skyblue')

critical_value = f.ppf(0.95, dfn, dfd)
plt.axvline(critical_value, color='red', linestyle='--', label='Critical value (95%)')

plt.title('F-Distribution Curve')
plt.xlabel('F-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

group1 = [23, 20, 25, 30, 22]
group2 = [35, 40, 38, 32, 37]
group3 = [45, 47, 50, 46, 48]

f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("There is a significant difference between the group means.")
else:
    print("No significant difference between the group means.")

data = [group1, group2, group3]
plt.boxplot(data, labels=["Group 1", "Group 2", "Group 3"])
plt.title("One-way ANOVA Group Comparison")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_1samp

np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=30)  # simulate data with mean=50, sd=5

t_stat, p_value = ttest_1samp(data, popmean=52)  # test if mean = 52

print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: mean is different from 52")
else:
    print("Fail to reject null: mean could be 52")
```

```python
import numpy as np
from scipy.stats import chi2

data = np.array([5, 7, 8, 6, 9, 7, 5, 6, 7, 8])
n = len(data)
sample_var = np.var(data, ddof=1)
pop_var = 4

chi_stat = (n - 1) * sample_var / pop_var
alpha = 0.05
df = n - 1
lower = chi2.ppf(alpha/2, df)
upper = chi2.ppf(1 - alpha/2, df)

print(f"Chi-square statistic: {chi_stat}")
print(f"Critical values: {lower}, {upper}")

if chi_stat < lower or chi_stat > upper:
    print("Reject null hypothesis: variance is different from", pop_var)
else:
    print("Fail to reject null: variance could be", pop_var)
```

```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 10, 10])

expected = np.array([25, 25, 25, 25])

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: observed data does NOT fit the expected distribution")
else:
    print("Fail to reject null hypothesis: observed data fits the expected distribution")
```
